## Bizantine Agreement in the Case of 4 Generals Coping with 2 Traitors:
## *A Quantum Solution*

To understand how the agreement is reached, consider an ideal system where, at this point, the lists of integers are:



            A fragment of lists was obtained by noise-free simulation, using the qasm simulator of Qiskit Aer.
            Successively, the list of Alice, Bob, Carol and Dave.
    
Alice separately and secretly sends the lieutenants an order, either retreat (0) or attack (1). The orders are the same, except if Alice is a traitor

The lieutenants inform each other of the orders received. They check if they are all identical. If so, they follow Alices order, and detectable broadcast is achieved.

Otherwise, there is at least one traitor among the generals. When two lieutenants report different orders, each must prove what she/he says. For this, they both embark on a turn-based game. Two games are played because all game scenarios include two pairs that signal different orders to each other.

For each game, the players draw whoever starts. They take turns sending an index whose corresponding entry in the opponents list must match the inverse of the order they said they received. Each player counts the errors. No index is used twice. The game ends when a player issues a stop flag indicating that her/his list of matching indexes is exhausted. And it still ends after a pre-agreed number of rounds (this justifies taking turns instead of swapping the lists in one move).

*Each of these scenarios achieves detectable broadcast:*

### No traitor:

Along with an order, Alice sends the list of corresponding indexes. Lets say Alice sends the order to attack (1) with the index list [0, 4, 14, 17, 18, 19, 20, 22, 27, 30, 31, 32, 33] .



No games is played.

###   Alice is the only traitor:
Alice sends conflicting orders, each with its index list. Lets say she sends attack (1) to Bob, the target of her cheating, and retreat (0) to Carol and Dave. These two receive as index list [7, 8, 10, 13, 16, 24, 26, 28, 34]

Bob vs. Carol and Bob vs. Dave:


                Alice orders Bob to attack (yellow) and orders Carol and Dave to retreat (cyan).
 
No error occurs when Bob and Carol are playing, nor when Bob and Dave are playing. The three lieutenants conclude that Alice is a traitor. They adopt the majority strategy.

###    A lieutenant is the only traitor:

Lets say the traitor is Bob and the order is attack (1). He has to face Carol and Dave.

For Bob, sending an index corresponding to a 0 in his list is a bad choice. With such an index present in the list sent by Alice, there is always an error. And in the remaining 2/3 of the cases, the index does not correspond at the same time to a 1 in the lists of Carol and Dave.

Bob, therefore, has no choice but to send indexes that correspond to 1 in his list, hoping that 1 is the corresponding entry in Carol and Daves lists.

Bob vs. Carol and Bob vs. Dave:


                 Bob sends indexes corresponding to 1 in his list. 
                 He succeeds 11 times against Carol and 12 times against Dave (green). 
                 He fails 4 times against Carol and 4 times against Dave ( purple). 
                 In 9 cases he succeeds and in 2 cases he fails against both Carol and Dave for the same index.
 
As long as the number of rounds is high, its unlikely that Bob wont make an error in either game. In the ideal case, the error rate will be 2/9 on average. Carol and Dave, unmasking Bob as a traitor, decide to follow Alices order. 
The same consensus occurs in the mirror cases where Carol or Dave is the sole traitor.

###    Alice and a lieutenant are traitors.

If Alice sends the same order to the loyal lieutenants and the reverse order to the traitor, all three lieutenants will say they received the same order. No game occurs and the two loyal lieutenants will follow the same strategy.
If Alice sends opposing orders to the loyal lieutenants, they will determine that they are not traitors during the game between them. One of them will play against the traitor who will be unmasked. In the end, loyal lieutenants adopt whatever strategy they find most reasonable when the source and an additional peer are unreliable.
